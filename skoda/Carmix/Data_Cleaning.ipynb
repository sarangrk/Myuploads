```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('epamc.csv', sep=';', encoding="ISO-8859-1")
```

```python
col_name = list(df.columns)
temp = []
for i in range(0,len(col_name)):
    temp.append(col_name[i].strip())
df.columns = temp
```
