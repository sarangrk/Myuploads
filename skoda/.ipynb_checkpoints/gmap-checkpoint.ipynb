### To plot routes on Map
Creating list of all the required address

```python
add_list = ['Teradata Polska Sp. z o.o., Wołoska 5, 00-001 Warszawa, Poland',
            'Skoda Auto Mladá Boleslav, tř. Václava Klementa 869, 293 01 Mladá Boleslav, Czechia']
# https://maps.googleapis.com/maps/api/geocode/json?address='%s'&key='%s'
```

For this we need to use __requests__ module of python. Where we will send the request to geomaps service api to featch the shortest path distance. We will get the distances in XML or JSON format. <br>
Initializing lists and other parameters for extraction.<br>

```python
results = []
location_latitude = []
location_longitude = []
formatted_address = []

for loc in add_list:
    add = loc.replace(" ","+")
    r = requests.get('https://maps.googleapis.com/maps/api/geocode/json?address='"%s"'&key=%s'%(add,key))
    result = r.json()['results']
    location = result[0]['geometry']['location']
    location_longitude.append(location['lng'])
    location_latitude.append(location['lat'])
    formatted_address.append(result[0]['formatted_address'])
    results.append(result)                             
```

```python
df = pd.DataFrame({'address':add_list,'latitue':location_latitude,'longitude':location_longitude,'formatted_address':formatted_address})
```

```python
source = (df.loc[0,'latitue'],df.loc[0,'longitude'])
destination = (df.loc[1,'latitue'],df.loc[1,'longitude'])
```

```python
from datetime import datetime
```

```python
gmaps.configure(api_key = key)
```

```python
now = datetime.now()
fig = gmaps.figure()
layer = gmaps.directions.Directions(source, destination, mode = 'car', api_key = key, departure_time = now)
fig.add_layer(layer)
fig
```
