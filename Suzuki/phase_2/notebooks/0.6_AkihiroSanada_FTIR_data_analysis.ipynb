    - 
- TYQ0810: 国内ワランティ明細情報
    - 

- ワランティ発生件数に効きそうな要因
    1. パーツに不具合があるか(FTIR発生/報告件数が多い欠陥品はワランティが多いだろう)
    2. パーツは壊れやすいか(地域, モデル, パーツ種類, 天候, 温度/湿度, 個人の使用状況, ドライバーの腕)
    3. 使用期間(生産からor購入からの期間、走行距離)
    4. 季節/月(繁忙期には車にたくさん乗るかもしれない)
    5. 主原因パーツの特定が進んだ(とりあえずこのパーツを交換しよう→交換数増加 / この不具合では、このパーツは影響しないことがわかった→交換数減少)
    6. パーツの市場流通数(ワゴンRの生産台数が同じでも、パーツが採用されなくなったら、市場に出回る数は減少)
    7. 代理店が交換しやすい(代金が高い、工数が小さいなど)
    8. 大規模なリコールがかかった

- 上記要因から考えられる、有効な特徴量
    1. 対象パーツのワランティ発生数, FTIR発生数, FPCR発生数, FTIR市場報告件数, パーツ生産数
    2. 対象パーツの使用されていた地域、搭載されていた車体モデルコード、当該地域の過去の気象条件(天候・温度・湿度)、当該車体の使用目的、駐車状況、主要走行道路舗装状況、運転者の年令、性別、ドライバー歴、ドライブ頻度
        - 車の扱いの荒いドライバーが多く乗る傾向のあるモデルは、ワランティが発生しやすいかもしれない
    3. 生産からの経過日数、購入からの経過日数、**前回点検からの経過日数、交換からの経過日数**、走行距離
    4. 故障発生日の月・週・曜日。ワランティ発行日の月・週・曜日
    5. 対象パーツの関連するFPCRの登録日、調査完了日、対策完了日、重要度
    6. 現在の対象パーツの流通数(生産データと回収データから算出
    7. 対象パーツの代金、交換時の平均工数
    8. 対象パーツの関連するリコール情報

- 上記の特徴量を作成するために必要なデータテーブル
    - 受領済みまたはTDで入手可能:
        - ワランティデータ(ワランティ発生数, 対象パーツの使用されていた地域, 搭載されていた車体モデルコード, 故障発生日の月・週・曜日。ワランティ発行日の月・週・曜日, 対象パーツの代金、交換時の平均工数)
        - 無料でダウンロード可能: 世界気象データ(当該地域の過去の気象条件(天候・温度・湿度))
    - 申出分析プロジェクトで受領したが、本プロジェクトでは未受領
        - FPCR基本情報(対象パーツの関連するFPCRの登録日、調査完了日、対策完了日、重要度, FPCR発生数)
        - FTIR基本情報(FTIR発生数)
        - FTIR市場報告(対象パーツの使用されていた地域、搭載されていた車体モデルコード、当該地域の過去の気象条件(天候・温度・湿度)、当該車体の使用目的、駐車状況、主要走行道路舗装状況、ドライブ頻度, FTIR市場報告件数)
        - コンポーネントコードマスター
    - もし可能であれば受領したい
        - モデルパーツマスター(生産データでは車両単位の生産数しかわからないため、車両とパーツを結びつけるテーブルがあると嬉しい)
        - 部品ニュースデータ(当該パーツに関連する部品ニュースの件数が知りたい。発表日時、対象部品、内容をレコードとするようなニュースデータがあると嬉しい)
        - リコール実施状況(リコール情報)
        - 整備履歴(故障していない車について、対象パーツの使用されていた地域、搭載されていた車体モデルコード、当該地域の過去の気象条件(天候・温度・湿度)、当該車体の使用目的、駐車状況、主要走行道路舗装状況、ドライブ頻度、走行距離)
        - 顧客情報データ(もし運転者の年令、性別、ドライバー歴、ドライブ頻度を含んだマスタデータがあれば。)

## アプローチ

1. !**時系列**!
    - 各パーツ毎に、毎月のワランティ数をARIMAX/状態空間モデルなどの時系列モデルでモデリング
    - $y$は各月のワランティ数($y_{1:t}$や$x_{1:t}$から$y_{t+1}$を予測)
    - その月をカテゴリ変数で入れれば、季節性も表現できる
    - 一部の共変量を組み込みづらいのが難点
        - 平均試用期間、各地域の割合などで組み込む

2. 生存分析
    - 各パーツ毎に、経過日数をtとして、セミパラメトリックな生存分析を行う
    - 生存関数$S_{x}(t)$を作成
    - 各パーツの生産情報と生存関数から、翌月のワランティ数を計算
    - 季節性が組み込みにくい。
        - 各VINの各パーツ毎に0,1でワランティ発生/発生なしが生成されるような階層ベイズモデルでモデリングすれば、季節性も組み込めるが、計算量が膨大になることが課題。

3. 機械学習1
    - 各パーツ毎にワランティ数を目的変数としてモデルする。1時系列アプローチと似ている。
    
4. 機械学習2
    - 各VIN毎の各パーツ毎に、翌月の故障0/1を予測する
    - 数が膨大になることが課題
    - アプローチに似ている
