```python
feature_name_dict = {key: val for _, (key, val) in df_feature_name.iterrows()}

df_CAN.feature_name = df_CAN.feature_name.apply(lambda x: feature_name_dict[x] if x in feature_name_dict.keys()
                                               else x)
```

```python
df_CAN_pivoted = df_CAN.pivot_table(index=["car_id","record_start_time","timestamp_sec"], columns="feature_name", 
                  values="feature_value")
```
