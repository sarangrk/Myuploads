## 使用列の絞り込み

```python
# データ取得可能年月日: F_REPORT_DATE
# 国内生産: F_MANUFACTURE_FACTORY_CODE == 100100
# 二輪/四輪: F_PRODUCT_SPECIFICATION in (1, 2)
# 主原因品番: F_MAIN_CAUSE_NO
# (再現性: F_FAULT_REPRODUCT)
# FTIR ID: F_ID (F_ID_ROOT?, F_ISSUE_NO?)
# F_SELLING_DATE, F_FAILURE_DATE
# 機種: F_SELLING_MODEL_SIGN

selected_col = ["F_ID", "F_VIN", "F_WRITING_DATE", "F_TEMP_REGIST_DATE",
             "F_REPORT_DATE", "F_MAIN_CAUSE_NO", "F_SELLING_DATE",
             "F_FAILURE_DATE", "F_SELLING_MODEL_SIGN"]

df_ftir_sub = (df_ftir.loc[df_ftir.F_PRODUCT_SPECIFICATION.isin([1, 2]) * \
             df_ftir.F_MANUFACTURE_FACTORY_CODE == "100100", 
                          selected_col])
```

```python
date_col = ["F_WRITING_DATE", "F_TEMP_REGIST_DATE", "F_REPORT_DATE",
            "F_SELLING_DATE", "F_FAILURE_DATE"]

for col in date_col:
    df_ftir_sub[col] = pd.to_datetime(df_ftir_sub[col].astype(str),
                                      errors="coerce")
```

```python
def dif_month(a, b):
    return 12 * (a.dt.year - b.dt.year) + (a.dt.month - b.dt.month) \
        + (a.dt.day - b.dt.day) / 31
```

```python
df_ftir_sub["m_sale_to_failure"] = np.round(
    dif_month(df_ftir_sub.F_FAILURE_DATE, df_ftir_sub.F_SELLING_DATE)
    ,0)
```

```python
import math
import matplotlib.pyplot as plt

def plot_hist(df, figsize=(15,10), subplot_col=3, bins=100, fontsize=20):
    s_num_list = [s for _, s in df.iteritems() if s.dtype.name != "object"]
    fig = plt.figure(figsize=figsize)
    k = len(s_num_list)
    for i, s in enumerate(s_num_list):
        ax = fig.add_subplot(math.ceil(k / subplot_col), subplot_col, i+1)
        s.hist(bins=bins)
        plt.title(s.name, fontsize=fontsize)
    plt.tight_layout()
```

```python
plot_hist(df_ftir_sub)
```

```python
def plot_category(df, figsize=(15,10), subplot_col=3, topN=10, fontsize=20,
                 include_other=True):
    s_num_list = [s for _, s in df.iteritems() if s.dtype.name == "object"]
    
    fig = plt.figure(figsize=figsize)
    k = len(s_num_list)
    for i, s in enumerate(s_num_list):
        vc = s.value_counts()
        vc_top = vc[:topN]
        if include_other: vc_top["other"] = vc[topN:].sum()
        ax = fig.add_subplot(math.ceil(k / subplot_col), subplot_col, i+1)
        vc_top.plot(kind="bar", fontsize=fontsize)
        plt.title(s.name, fontsize=fontsize)
    plt.tight_layout()
```

```python
plot_category(df_ftir_sub, subplot_col=4, include_other=False, figsize=(19,5))
```

```python
def plot_pareto(df, figsize=(15,10), subplot_col=3, fontsize=20,
                percentile=0.8):
    s_num_list = [s for _, s in df.iteritems() if s.dtype.name == "object"]
    
    fig = plt.figure(figsize=figsize)
    k = len(s_num_list)
    for i, s in enumerate(s_num_list):
        vc = s.value_counts()
        cover_ratio = vc.cumsum() / vc.sum()
        annot_x = np.arange(len(vc))[cover_ratio >= percentile][0]
        annot_y = cover_ratio[annot_x]
        ax = fig.add_subplot(math.ceil(k / subplot_col), subplot_col, i+1)
        plt.plot(cover_ratio.values)
        ax.annotate(f"q={percentile}: x={annot_x}", xy = (annot_x, annot_y),
                    size = fontsize)
        plt.plot([annot_x], [annot_y], "o")
        plt.title(s.name, fontsize=fontsize)
    plt.tight_layout()
```

```python
plot_pareto(df_ftir_sub, subplot_col=4, figsize=(19,4))
```

## FTIR件数の時系列

```python
df_ftir_sub["report_month"] = \
    pd.to_datetime(df_ftir_sub.F_REPORT_DATE.dt.strftime("%Y%m") + "01")
ftir_monthly = (df_ftir_sub
                .groupby(["F_MAIN_CAUSE_NO","report_month"])["F_ID"]
                .agg({"n":len}))
```

```python
ftir_monthly.head()
```

```python
ftir_monthly.shape
```

```python
ftir_part_count = (ftir_monthly
                     .reset_index()
                     .groupby("F_MAIN_CAUSE_NO")["n"]
                     .sum()
                     .sort_values(ascending=False))
```

```python
ftir_part_count.head()
```

```python
fig = plt.figure(figsize=(30,15))
for i, part_no in enumerate(ftir_part_count.index[:42]):
    ax = fig.add_subplot(6, 7, i+1)
    tmp = ftir_monthly.loc[part_no]
    ax.plot(tmp.index, tmp.values, "o-")
    ax.set_xlim(datetime(2007,1,1), datetime(2018,12,31))
    ax.set_title(part_no)
plt.tight_layout()
plt.show()
```

```python
ftir_weekly = (df_ftir_sub
                .groupby(["F_MAIN_CAUSE_NO",pd.Grouper(key="F_REPORT_DATE", freq="W-MON")])["F_ID"]
                .agg({"n":len}))
ftir_weekly.reset_index(inplace=True)
ftir_weekly["F_REPORT_DATE"] = ftir_weekly["F_REPORT_DATE"] - pd.to_timedelta(7, unit='d')
ftir_weekly.rename(columns={"F_REPORT_DATE":"F_REPORT_WEEK"}, inplace=True)
ftir_weekly.set_index(["F_MAIN_CAUSE_NO", "F_REPORT_WEEK"], inplace=True)
```
